# 3. Simulating the Stock Market

**Let's take a look at a classic finance problem.** How could we model a stock, or any changing asset? Despite sounding very vague and possibly very challenging, there is a common approach that can actually be very straight-forward to model: it's called random walk.


**Random walk** is the approach that each change is completely random, meaning it doesn't depend on any previous changes (for my stats people, think *independent*). 


### So, what does all this mean?
Basically, we want to plot a stock's value (for example, Google's stock value) over time. Maybe we want to plot what we think the value will look like over the next 50 days. What we'll need is a *return* value. Returns (say the random variable $R$) follow the normal distribution below:

$R$ ~ $N(\frac{\mu}{T}, \frac{v}{\sqrt{T}})$, where $\mu$ is the mean returns of the asset, $v$ is the volatility of the asset, and $T$ is the total time you want to predict for.

How the heck does that thing above translate to code? Remember that we can get random values from numpy, ```np.random```. Well it also has distributions like the normal distribution, uniform distribution, etc. The documentation is [here](https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.normal.html). We should use the average returns of the stock, $\mu$, and the volatility of the stock, $v$, in this function call ```np.random.normal(mean, standard_deviation)```.
